```r
p1=dbGetQuery(con,'select Player.player_name, Player.player_api_id,max(Player_Attributes.overall_rating)AS overall_rating,(max(Player_Attributes.gk_diving)+max(Player_Attributes.gk_handling)+max(Player_Attributes.gk_kicking)+max(Player_Attributes.gk_positioning)+max(Player_Attributes.gk_reflexes)) AS gk_ratings from Player, Player_Attributes where Player.player_api_id=Player_Attributes.player_api_id Group by Player.player_api_id')
```
